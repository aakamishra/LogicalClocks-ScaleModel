```python
dfs = {}
for i in range(1, 4):
    df = pd.read_csv(f"logs/client-{i}.csv", index_col=False)
    df.rename(columns={df.columns[1]: 'System Time'}, inplace=True)
    df = df.iloc[1:]
    dfs[i] = df
```

```python
start_time = min([x['System Time'].iloc[0] for x in dfs.values()])
print("Start Time", start_time)

sends = (dfs[1][(dfs[1]['Event Type'] == "SEND") & (dfs[1]['Reciever'] == '2')]["System Time"] - start_time).tolist()
receives = (dfs[2][(dfs[2]['Event Type'] == "RECEIVED") & (dfs[2]['Sender'] == '1')]["System Time"] - start_time).tolist()

sends = sends[:len(receives)]

length = len(receives)

for _ in range(length):

plt.plot(1 * np.ones(length), sends, 2 * np.ones(length), receives)

# for i in range(1, 4):
#     plt.scatter(i * np.ones(len(dfs[i])), dfs[i]["System Time"] - start_time)

```
